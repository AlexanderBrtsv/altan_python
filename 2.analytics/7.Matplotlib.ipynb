# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

- простые линейные графики

```python
# 1-мерный массив из 10 значений
x = np.arange(10)
x
```

```python
y = x * 2
y
```

```python
plt.plot(x, y)
plt.show()
```

```python
# 1-мерный массив из 1000 значений 
# линейная последовательность в диапазоне от 0 до 10
x = np.linspace(0, 10, 1000)
x
```

```python
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-10, 10, 10)
y = np.tanh(x)
plt.plot(x, y)
plt.show()
```

```python
#def sigmoid(x):
#    return 1 / (1+np.exp(-x))

sigmoid = lambda a: 1 / (1+np.exp(-a))
y = sigmoid(x)

plt.plot(x, y)
plt.show()
```

```python
# совмещение двух графиков в одном окне
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.cos(x -1)

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

```python
# настройка графиков: цвет и вид линий
# цвет можно задавать разными способами
plt.plot(x, np.cos(x), color="red") # по названию

plt.plot(x, np.cos(x+1), color="y") # по первой букве цвета из rgbcmyk

plt.plot(x, np.cos(x - 1), color="#555555") # по 16-ричному коду RGB

plt.plot(x, np.cos(x + 2), color=(0.5, 1.0, 0.8)) # RGB-кортеж со значениями от 0 до 1
plt.show()
```

```python
# стиль линий
plt.plot(x, np.sin(x), linestyle="-") # сплошная
#plt.plot(x, np.sin(x+1), linestyle="solid") # сплошная
plt.plot(x, np.sin(x-1), linestyle="--") # штриховая
#plt.plot(x, np.sin(x-2), linestyle="dashed") # штриховая
plt.plot(x, np.sin(x-4), linestyle=":") # пунктирная
plt.plot(x, np.sin(x+4), linestyle="-.") # штрих-пунктирная
plt.show()
```

```python
# объединение цветов и стилей линий
plt.plot(x, np.sin(x), linestyle="-", color="red") # сплошная

plt.plot(x, np.sin(x-2), linestyle="--", color="k") # штриховая

plt.plot(x, np.sin(x-4), linestyle=":", color="g") # пунктирная

plt.show()
```

```python
# объединение цветов и стилей линий
plt.plot(x, np.sin(x), "-g") # сплошная

plt.plot(x, np.sin(x-2), "--k") # штриховая

plt.plot(x, np.sin(x-4), ":b") # пунктирная

plt.show()
```

```python
# пределы осей
plt.plot(x, np.sin(x), "-g")

# индивидуальные настройки осей
#plt.xlim(0, 10)
#plt.ylim(-0.75, 0.75)

plt.axis([-15, 15, -1.5, 1.5])
plt.show()
```

```python
# легенда графика 
plt.plot(x, np.sin(x), label = "синусоида")
plt.plot(x, np.cos(x), label = "косинусоида")

plt.legend()
plt.axis([-20, 20, -2.5, 2.5])
plt.show()
```

```python
# подписи осей 
plt.plot(x, np.sin(x), label = "sin")
plt.plot(x, np.cos(x), label = "cos")

plt.legend()
plt.xlabel("x")
plt.ylabel("sin and cos")
plt.axis([-20, 20, -2.5, 2.5])
plt.show()
```

```python
# заголовок графиков 
plt.plot(x, np.sin(x), label = "sin")
plt.plot(x, np.cos(x), label = "cos")

plt.legend()
plt.xlabel("x")
plt.ylabel("sin and cos")
plt.title("Графики sin и cos")
plt.axis([-20, 20, -2.5, 2.5])
plt.show()
```
