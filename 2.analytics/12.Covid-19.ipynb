# Анализ данных по COVID-19

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = pd.read_csv("data/covid_19_data.csv")
data_1.head(10)
```

```python
data_1.shape
```

```python
data_2 = data_1.drop(["SNo", "Last Update"], axis = 1)
data_2.head(10)
```

```python
# проверка на дубликат
data_2.duplicated(["ObservationDate", "Province/State", "Country/Region"]).sum()
```

```python
# удаление дубликатов
data_3 = data_2.drop_duplicates(["ObservationDate", "Province/State", "Country/Region"])
```

```python
data_3.duplicated(["ObservationDate", "Province/State", "Country/Region"]).sum()
```

```python
data_3["Country/Region"].unique()
```

```python
data_3["Country/Region"].str.strip(" ").unique()
```

```python
# исправляем названия стран
data_3.loc[:,"Country/Region"] = data_3.loc[:,"Country/Region"].str.strip(" (),''")#.unique()
```

```python
data_3.loc[:,"Country/Region"].unique()
```

```python
data_3.columns
```

```python
# изучаем страну "Others"
data_3[data_3["Country/Region"] == "Others"]
```

```python
# конвертация временных данных в тип TimeStamp
data_3.loc[:,'ObservationDate'] = pd.to_datetime(data_3['ObservationDate'])
```

```python
# извлечение дат из TimeStamp
data_3.loc[:,'ObservationDate'] = data_3['ObservationDate'].apply(lambda el: el.date())
```

```python
# анализ
# группируем данные по датам и выводим суммарные значения(по всему миру)
data_3.groupby(["ObservationDate"]).sum()
```

```python
data_3.columns
```

```python
# статистика по странам и датам

data_country = data_3.groupby(["Country/Region", "ObservationDate"])[['Confirmed','Deaths','Recovered']].sum()
data_country.head(10)
```

```python
# данные максимумов по странам
data_max = data_country.groupby(["Country/Region"]).max()
data_max.head(10)
```

```python
# выбираем страну или страны
data_max.loc["Russia", :]
```

```python
data_date = data_3.groupby(["ObservationDate"]).sum()
data_date.head(10)
```

```python
plt.figure(figsize=[10, 5])

plt.yscale("log")

plt.plot(data_date.index, data_date["Confirmed"], label="Заболевшие")

plt.plot(data_date.index, data_date["Deaths"], label="Погибшие")

plt.plot(data_date.index, data_date["Recovered"], label="Выздоровевшие")
plt.legend()
plt.show()
```

```python
# оценка летальности
plt.figure(figsize=[10, 5])

letal_data = (data_date["Deaths"] / data_date["Confirmed"]) * 100

plt.plot(data_date.index, letal_data)

plt.show()
```

```python
# новые случаи заболевания

# функция для вычисления новых случаев
def new_cases_calculate(df, col_idx):
    new_cases = []
    for idx in range(df.shape[0]):
        if idx == 0:
            new_cases.append(df.iloc[idx, col_idx])
        else:
            new_cases.append(df.iloc[idx, col_idx] - df.iloc[idx-1, col_idx])
    return new_cases
```

```python
new_cases_calculate(data_date, "Confirmed")
```

```python
plt.figure(figsize=[10, 5])

plt.plot(data_date.index, new_cases_calculate(data_date, "Confirmed"), label="Новые случаи заболевания")
# plt.plot(data_date.index, data_date["confirmed"], label="Заболевшие")
plt.legend()
plt.show()
```
