- Диаграммы рассеивания

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.scatter(x, y, marker="$4$", s=50)
plt.show()
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)

# задаем размеры каждой точки массивам (размер массива должен соответствовать размеру массива данных)
sizes = np.random.randint(10, 100, 50)

# задаем цвета каждой точки массивам (размер массива должен соответствовать размеру массива данных)
colors = np.linspace(0, 1, 50)

plt.scatter(x, y, marker="*", s=sizes, c = colors, cmap = "viridis")
plt.show()
```

- пример с городами Якутии

```python
import pandas as pd
```

```python
p = {
        "Yakutsk": 318768,
        "Neryungri": 57009,
        "Viluysk": 11095
}

a = {
        "Yakutsk": 122,
        "Viluysk": 15,
        "Neryungri": 98.9
}

# координаты (lat - широтаб lon - долгота)    
lat = {
    "Yakutsk": 62,
    "Neryungri": 56,
    "Viluysk": 63
}

lon = {
    "Yakutsk": 129,
    "Neryungri": 124,
    "Viluysk": 121
}
    
data = pd.DataFrame(
    {"population": p, "area": a, "latitude": lat, "longitude": lon}
)
     
data
```

```python
# визуализация данных

# логарифмирование данных по населению
pop_log = np.log10(data["population"])
plt.scatter(data["longitude"], data["latitude"], s=data["area"], c=pop_log)

# цветовая шкала
plt.colorbar(label="log10(население)")
plt.xlabel("dolgota")
plt.ylabel("shirota")

plt.show()
```

- объединение графика и диаграммы рассеивания

```python
x = np.linspace(0, 10, 1000)
y = np.sin(x)

# модуляция для scatter
z = y + np.random.randn(1000) * 0.2

# массив цветов
#c = np.random.randint(0, 10, size=1000)
c = np.linspace(0, 10, 1000)

# размер точек 
s = np.linspace(0, 20, 1000)

# график
plt.plot(x, y)

#диаграмма рассеивания
plt.scatter(x, z, c=c, alpha=0.5, s=s, cmap="plasma")

plt.show()
```
