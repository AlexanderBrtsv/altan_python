# Работа с Excel

- создание из датафрейма

```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("titanic.csv")
data.head()
```

```python
# группировка данных по "survived", "pclass", "who" и узнаем количество человек
```

```python
count_data = data.groupby(["survived", "pclass", "who"])["who"].count()
count_data
```

```python
# группировка данных по "survived", "pclass", "who" и узнаем средний возраст
```

```python
mean_data = data.groupby(["survived", "pclass", "who"])["age"].mean()
mean_data
```

```python
df = pd.DataFrame({"count":count_data, "mean":mean_data})
df
```

```python
# создание Excel-файла
```

```python
df.to_excel("titanic1.xlsx")
```

```python
# создание Excel-файла с названием листа
```

```python
df.to_excel("titanic1.xlsx", sheet_name = "Titanic")
```

```python
# создание Excel-файла с несколькими листами
```

```python
# словарь с датафреймами
```

```python
#sheets = {"Titanic_1": data, "Titanic_2": df}

# определение движка для записи нескольких датафреймов
excelWriter = pd.ExcelWriter("titanic2.xlsx", engine = "xlsxwriter")

excelWriter
```
